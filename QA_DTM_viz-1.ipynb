```python
import pandas as pd
import plotly.express as px
```

```python
qa_dtm_td = pd.read_csv("/Users/siradachuntararuangnapa/Desktop/DTM/QA_DTM_td_all_time.csv")
qa_dtm_wt = pd.read_csv("/Users/siradachuntararuangnapa/Desktop/DTM/QA_DTM_wt_all_time.csv")
```

## QA: All Time Period Common Topics

```python
fig = px.line(qa_dtm_td, x="timepoint", y=qa_dtm_td.columns,
              title='QA: All Time Period Common Topics')
fig.update_xaxes(
    dtick="M1",
    tickformat="%b %d",
    ticklabelmode="period")
fig.show()
```

## QA: Early Time Period Common Topics 1 (Year 2011-2012)

```python
fig = px.line(qa_dtm_td, x="timepoint", y= ['topic_3', 'topic_4', 'topic_9'],
              title='QA: Early Time Period Common Topics 1 (Year 2011-2012)')
fig.update_xaxes(
    dtick="M1",
    tickformat="%b %d",
    ticklabelmode="period")
fig.show()
```

```python
# topic_3
topic_3 = qa_dtm_wt[qa_dtm_wt['k']==2] # start with 0
# select timepoint
topic_3_3 = topic_3[topic_3['timepoint']==3]
# concat 
topic_3_concat = topic_3_3.copy()
topic_3_concat.sort_values(by='prob', ascending=False, inplace=True)
#print(topic_3_concat.T)

# topic_4
topic_4 = qa_dtm_wt[qa_dtm_wt['k']==3] # start with 0
# select timepoint
topic_4_4 = topic_4[topic_4['timepoint']==4]
# concat 
topic_4_concat = topic_4_4.copy()
topic_4_concat.sort_values(by='prob', ascending=False, inplace=True)
#print(topic_4_concat.T)

# topic_9
topic_9 = qa_dtm_wt[qa_dtm_wt['k']==8] # start with 0
# select timepoint
topic_9_6 = topic_9[topic_9['timepoint']==6]
# concat 
topic_9_concat = topic_9_6.copy()
topic_9_concat.sort_values(by='prob', ascending=False, inplace=True)
#print(topic_9_concat.T)

# concat all together
topic_all_early1 = pd.concat([topic_3_concat,topic_4_concat, topic_9_concat], ignore_index=True)

# groupby word and get the average of prob
topic_all_early1 = topic_all_early1.groupby(['word', 'timepoint'])['prob'].mean().reset_index()
# sort by prob
topic_all_early1.sort_values(by='prob', ascending=False, inplace=True)
topic_all_early1.T
```
